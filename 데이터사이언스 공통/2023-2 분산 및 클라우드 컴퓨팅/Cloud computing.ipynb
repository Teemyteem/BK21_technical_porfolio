```python
#Replace outlier 
def replace_outlier_sd_mean(val, mean, std):
    if val > mean + (3*std):
        return mean 
    elif val < mean - (3*std):
        return mean 
    return val

def replace_outlier_sd_median(val, mean, median, std):
    if val > mean + (3*std):
        return median
    elif val < mean - (3*std):
        return median
    return val

def replace_outlier_sd_mode(val, mean, mode, std):
    if val > mean + (3*std):
        return mode
    elif val < mean - (3*std):
        return mode
    return val

def replace_outlier_iqr_mean(val, mean, q1, q3, IQR):
    if val > q3+1.5*IQR:
        return mean
    elif val < q1-1.5*IQR:
        return mean
    return val

def replace_outlier_iqr_median(val, median, q1, q3, IQR):
    if val > q3+1.5*IQR:
        return median
    elif val < q1-1.5*IQR:
        return median
    return val

def replace_outlier_iqr_mode(val, mode, q1, q3, IQR):
    if val > q3+1.5*IQR:
        return mode
    elif val < q1-1.5*IQR:
        return mode
    return val

def replace_outlier_sd_indicator(val, mean, std):
    if val > mean + (3*std):
        return True
    elif val < mean - (3*std):
        return True
    else:
        return False
    return val

def replace_outlier_iqr_indicator(val, q1, q3, IQR):
    if val > q3+1.5*IQR:
        return True
    elif val < q1-1.5*IQR:
        return True
    else:
        return False
    return val
```

```python
#Clean dataset
#Train
for col in df_train.columns:
    if col not in ['LocationName', 'Rating', 'latitude', 'longitude', 'zipcode']:
        mean = df_train[col].mean()
        median = df_train[col].median()
        mode = df_train[col].mode()
        std_dev = df_train[col].std(axis=0)
        
        q1 = df_train[col].quantile(0.25)
        q3 = df_train[col].quantile(0.75)
        IQR = q3-q1
        
        #SD
        df_train_sd_mean[col] = df_train[col].map(lambda x: replace_outlier_sd_mean(x, mean, std_dev))  
        df_train_sd_median[col] = df_train[col].map(lambda x: replace_outlier_sd_median(x, mean, median, std_dev))  
        df_train_sd_mode[col] = df_train[col].map(lambda x: replace_outlier_sd_mode(x, mean, mode[0], std_dev))
        
        df_train_sd_indicator[col] = df_train[col].map(lambda x: replace_outlier_sd_indicator(x, mean, std_dev))  
        
        #IQR
        df_train_iqr_mean[col] = df_train[col].map(lambda x: replace_outlier_iqr_mean(x, mean, q1, q3, IQR))    
        df_train_iqr_median[col] = df_train[col].map(lambda x: replace_outlier_iqr_median(x, median, q1, q3, IQR)) 
        df_train_iqr_mode[col] = df_train[col].map(lambda x: replace_outlier_iqr_mode(x, mode[0], q1, q3, IQR))
        
        df_train_iqr_indicator[col] = df_train[col].map(lambda x: replace_outlier_iqr_indicator(x, q1, q3, IQR)) 
```
